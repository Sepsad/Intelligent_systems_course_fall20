<h1 dir="rtl">الف
<h3 dir="rtl"> پاداشی  که دقیقا پس از انجام یک تصمیم در محیط حالت دریافت می شود را پاداش لحظه‌ای (immediate reward) می‌گویند. صرفا با استفاده از پاداش فوری،نمی‌توان تصمیماتی بهینه برای رسیدن به حالت ایده‌آل را به درستی انجام داد، چون در یادگیری تقویتی می‌خواهیم ربات رادر بلند مدت بهینه کنیم. به عنوان مثال ممکن است لازم باشد تصمیمی با پاداش لحظه‌ای منفی گرفته شود، در حالی که این تصمیم در آینده منجر به پاداش مثبت زیادی شود.
<h3 dir="rtl"> برای همین قضیه پاداش موخر (delayed reward) تعریف می شود، که حالات آینده را نیز در محاسبه پاداش در نظر می گیرد. پاداش موخر با استفاده از (discount factor) محاسبه می‌شود که مقدار این پارامتر میان ۰ و ۱ است. مقادیر نزدیکتر به ۱، سیستم را به سمت پاداش‌های آینده می‌برد، در حالی که مقادیر پایین‌تر سیستم را به سمت پاداش‌های لحظه‌ای می‌برد.

<h1 dir="rtl">ب
<h3 dir="rtl"> وقتی مسئله ما real-world است و بدون مدل از پیش تعیین شده است، روش های مستقل به جای روش های مبتنی بر مدل کاربرد بیشتری خواهند داشت. چون وقتی این مسئله‌ها از مدل ها برای یادگیری و تصمیم‌گیری استفاده می شود، خطاهای مدل در شبیه سازی می‌تواند در طول زمان باعث ناپایداری و بی ثباتی شود، زیرا این خطاها هرچه یادگیری جلوتر برود در هم ضرب شده و بزرگ می شوند. بدین ترتیب نتیجه به دست آمده در طول فاز planning که مبتنی بر شبیه سازی مدل است در این حالت کمتر قابل اعتماد است.

# Q2

```python
import pandas as pd
import numpy as np
from env import ENV 
from Q_table import Q_table
import matplotlib.pyplot as plt

states = pd.read_csv("ENV.csv", header= None)
```

# For R2 = 0

```python
REWARD_DICT = { 0:-0.01, 1:0.0, 2:-100, 3:100, 4:0, 5:-20}

enviorment = ENV(states = states, reward_dict = REWARD_DICT, random_starting_point = False)

q_table = Q_table(env = enviorment, gamma=0.9, n = 15, alpha=0.1, num_episodes=3000)

total_reward_dict_0 = q_table.learn( 'Q_table_for_R2_0' ,random_starting_point = True)
```

```python
q_table.test(random_starting_point = True)
```

```python
plt.figure(figsize= (30,5))
plt.plot(list(total_reward_dict_0.keys()), (list((total_reward_dict_0.values()))))
```

# For R2 = 30

```python
enviorment = ENV(states = states, reward_dict = { 0:-0.01, 1:0, 2:-100, 3:100, 4: 30, 5:-20}, random_starting_point = True)

q_table = Q_table(env = enviorment, gamma=0.9, n = 15, alpha= 0.1, num_episodes=3000)

total_reward_dict_30 = q_table.learn(Q_table_name  = 'Q_table_for_R2_30' ,random_starting_point = True)
```

```python
q_table.test(random_starting_point = False)
```

```python
plt.figure(figsize= (30,5))
plt.plot(list(total_reward_dict_30.keys()), (list((total_reward_dict_30.values()))))
```

<h3 dir="rtl">greedy سیاستی است که بهترین اقدام (یعنی عملی که با بالاترین مقدار همراه است) را با احتمال   1-eو یک اقدام تصادفی با احتمال  eرا انتخاب می کند. مشکلی که در مورد e-greedy وجود دارد این است که ، وقتی اقدامات تصادفی را انتخاب می کند (یعنی با احتمال زیاد ...) ، آنها را یکنواخت انتخاب می کند (یعنی همه اقدامات را به همان اندازه خوب می داند) ، حتی اگر برخی اقدامات (به استثنای بهترین فعلی) بهتر باشند از دیگران. در مواردی که اقدامات خاصی بسیار بدتر از سایر اقدامات باشد ، این روش ایده آل نیست. بنابراین ، یک راه حل طبیعی برای این مسئله انتخاب اقدامات تصادفی با احتمال متناسب با مقادیر فعلی آنها است. به این سیاست ها سیاست های softmax گفته می شود.
