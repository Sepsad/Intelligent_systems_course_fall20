```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from random import randrange
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
def calc_entropy(labels):
    freq = labels.value_counts()
    prob_classes = np.array(freq / freq.sum())
    entropy = 0
    for prob in prob_classes:
        if (prob == 0):
            continue
        entropy += -prob*np.log(prob)
    return entropy




class Node(object):
    def __init__(self, index = None, children = [], entropy = 0, depth = 0):
        self.index = index
        self.children = children
        self.entropy = entropy
        self.depth = depth
        
        # if Node isn't leaf
        self.split_feature = None
        self.order = None
        
        # if Node is leaf
        self.label = None
    
    def set_properties(self, split_feature, order):
        self.split_feature = split_feature
        self.order = order
    
    def set_label(self, label):
        self.label = label
        

class DecisionTree(object):
    
    def __init__(self, X = None , Y = None , max_depth = 3, min_sample_split = 3, min_I_gain = 1e-5):
        
        self.x = X
        self.features = list(X.columns)
        self.y = Y
        self.labels = Y.unique()
        self.train_number = X.shape[0]
        
        self.root_node = None
        
        self.max_depth = max_depth
        self.min_sample_split = min_sample_split
        self.min_I_gain = min_I_gain
    
        
        
    def entropy(self, indexes):
        
        if(len(indexes) == 0):
            return 0
        
        labels_of_node = self.y[indexes]
        return calc_entropy(labels_of_node)
    
    def split_node(self, node):
        
        indexes = node.index
        best_I_gain = 0
        best_feature = None
        best_splits = []
        order = None
        
        data_of_node = self.x.iloc[indexes]
        
        for i, feature in enumerate(self.features):
            values = list(self.x[feature][indexes].unique())
            
            # entropy is 0
            if(len(values) == 1):
                continue
            
            splits = []
            for value in values:
                splitted_indexes = list(data_of_node.index[data_of_node[feature] == value])
                splits.append(splitted_indexes)
            
            min_split = min(map(len, splits))
            
            # stop splitting for small size nodes
            if(min_split < self.min_sample_split):
                continue
            
            mean_entropies = 0
            
            for split in splits:  
                mean_entropies += (float(len(split)) / len(indexes)) * self.entropy(split)
            
            information_gain = node.entropy - mean_entropies
            
            # continue for small gain
            if (information_gain < self.min_I_gain):
                continue
            
            if (information_gain > best_I_gain):
                
                best_I_gain = information_gain
                best_feature = feature
                best_splits = splits
                order = values
        node.set_properties(best_feature, order)
        
        childs = []
        
        for split in best_splits:
            child_node = Node(index = split, entropy = self.entropy(split), depth = node.depth + 1)
            childs.append(child_node)
            
        return childs
    
    def set_label(self, node):
        votes = self.y[node.index]
        majority_vote = votes.mode()
        if(len(majority_vote) > 1):
            majority_vote = majority_vote[0]
        
        node.set_label(majority_vote)
        
    def fit(self):
        
        all_index = list(range(self.train_number))
        self.root = Node(index= all_index, children=[], entropy= self.entropy(all_index), depth= 0)
        nodes_of_tree = [self.root]
        
        while nodes_of_tree:
            
            node = nodes_of_tree.pop()
            
            if ((node.depth < self.max_depth) or (node.entropy < self.min_I_gain)):
                node.children = self.split_node(node)
                
                if (not node.children):
                    self.set_label(node)
                    
                    
                nodes_of_tree += node.children

            else:
                self.set_label(node)
    
    def predict(self, x_test):
        
        number_of_datas = x_test.shape[0]
        labels = []
        
        for i in range(number_of_datas):
            
            row = x_test.iloc[i , :]
            
            node = self.root
            while(node.children):
                feature = node.split_feature
                order_of_feature = row[feature]
                index_of_children = node.order.index(order_of_feature)
                node = node.children[index_of_children]
        
            labels.append(node.label[0])
            
        return (pd.Series(labels))


    
class RandomForest(object):
    
    def __init__(self,  data = None , target_feature = None, number_of_trees = 10, number_of_feature = 'sqrt', bootstrap_fraction = 0.9  ,max_depth = 3, min_sample_split = 3):
        
        self.data = data
        self.target = target_feature
        
        if number_of_feature == 'sqrt':
            self.n_features = int(np.sqrt(data.shape[1] -1))
        else:
            if(number_of_feature > (data.shape[1] -1)):
                self.n_features = (data.shape[1] -1)
            else:
                self.n_features = number_of_feature
        
        
        self.bootstrap = bootstrap_fraction
        self.n_trees = number_of_trees
        
        self.max_depth = max_depth
        self.min_sample_split = min_sample_split
        
    def fit(self):
        
        self.forest = []
        
        for i in range(self.n_trees + 1):
            train_data = self.data.sample(frac= self.bootstrap, replace=True)
            
            x_bootstrapped = train_data.drop(columns= [self.target]).reset_index(drop = True)
            y_bootstrapped = train_data[self.target].reset_index(drop = True)
        
            tree = DecisionTree(X= x_bootstrapped, Y= y_bootstrapped, max_depth= self.max_depth, min_sample_split= self.min_sample_split)
            
            tree.fit()
            Y_pred = Tree.predict(x_bootstrapped)

            print("\nAccuracy Score of tree {d} : ".format(d = i+1), accuracy_score(y_bootstrapped, Y_pred))
            
            
            
            self.forest.append(tree)
            
    def predict(self, x_test):
        
        number_of_datas = x_test.shape[0]
        labels = []
        
        for i in range(number_of_datas):
        
            row = x_test.iloc[[i]]
            votes = []
            
            for tree in self.forest:
                
                vote = tree.predict(row)
                votes.append(vote[0])
            
            votes = pd.Series(votes)
            labels.append(votes.mode()[0])
        
        return (pd.Series(labels))
            
def test_train_split(data, test_percent = 0.2, seed = 42):
    
    df = data.sample(frac=1, random_state= seed, replace= False).reset_index(drop=True)
    train_size = int((1 - test_percent) * df.shape[0])
    
    test_data =  df.iloc[train_size : ,:].reset_index(drop=True)
    train_data = df.iloc[ : train_size,:].reset_index(drop=True)
    return train_data, test_data
```

```python
dictionary = pd.read_csv(path_datests.format(d = 'dictionary.csv'))
coded_words = pd.read_csv(path_datests.format(d = 'coded_words.csv'))
```

```python
CHARS = 'abcdefghijklmnopqrstuvwxyz'

DICTIONARY = set(dictionary['word'])

CODED_WORDS = list(coded_words['word'])

POPULATION_SIZE = 100

TOP_POPULATION = 20

REMAIN_PERCENT = 0.1

CROSSOVER_PERCENT = 0.2

CROSSOVER_SIZE = int(len(CHARS) / 3)

MUTATIONS_PERCENT = 0.05

MUTATIONS_SIZE_PERCENT = 0.1

STABILITY_INTERVALS = 200
```

### پیاده‌سازی جمعیت و کراس‌اور و جهش



</div>

```python
def init_mapping():
    keys = set(CHARS)
    mapping = {}
    for c in CHARS:
        key = random.choice(list(keys))
        keys.remove(key)
        mapping[c] = key
    return mapping

def flip_dict(ini_dict):
    flipped = dict()
  
    for key, value in ini_dict.items(): 
        if value not in flipped: 
            flipped[value] = [key] 
        else: 
            flipped[value].append(key)
    
    return flipped

def correct_duplicate_value(child):
    
    if(not (set(child.keys()) - set(child.values()))):
        
        return child
    
    child_flipped = flip_dict(child)
    remaining_chars = list(set(CHARS) - set(child.values()))

    for k,v in child_flipped.items():
        
        if(len(v) > 1):
            v  = list(v)
            random_item_from_v = random.choice(v)
            v.remove(random_item_from_v)

            for key in v:
                random_item_from_remaining_chars = random.choice(remaining_chars)
                child[key] = random_item_from_remaining_chars
                remaining_chars.remove(random_item_from_remaining_chars)
    
    return child
"""""    
def crossover1(x, y, chars):
    
    for char in chars:
        
        value_x = x[char]
        value_y = y[char]

        y[char] = value_x
        x[char] = value_y
    
    x = correct_duplicate_value(x)
    y = correct_duplicate_value(y)
    
    return x, y
"""""

def crossover2(child1, child2, break_index = int(len(CHARS) / 2)):
    
    child1_items = list(child1.items())
    child2_items = list(child2.items())
    
    child1_first_part = child1_items[ : break_index]
    child1_second_part = child2_items[break_index :]
    child2_first_part = child1_items[ : break_index]
    child2_second_part = child2_items[break_index :]
    
    new_child1 = dict(child2_first_part + child1_second_part)
    new_child2 = dict(child1_first_part + child2_second_part)
    
    new_child1 = correct_duplicate_value(new_child1)
    new_child2 = correct_duplicate_value(new_child2)
    
    return new_child1, new_child2

def mutate(child, char1, char2):
    
    val1 = child[char1]
    val2 = child[char2]
    
    child[char1] = val2
    child[char2] = val1
    
    return child
    

def decode_word(word, mapping):
    result = ""
    for w in word:
        result += mapping[w]
    return result

def fitness_function(mapping):
    
    score = 0.0
    
    for word in CODED_WORDS:
        
        if( decode_word(word,mapping) in DICTIONARY ):
            score +=1

    return (score )
```

<div dir="rtl">
---------------------------
    
    
### پیاده‌سازی روتین‌های الگوریتم ژنتیک



</div>

```python
def select(population):
    scores = []
    
    for p in population:
        scores.append((fitness_function(p), p))

    sorted_population = sorted(scores, reverse=True, key=lambda tup: tup[0])

    selected_population = sorted_population[:TOP_POPULATION]
    best_score = selected_population[0][0]
    top_population = [m for _, m in selected_population]
    
    return best_score, top_population


def generate(population):
    
    new_population = population[:2]
#     new_population = population[:]

    
    while len(new_population) < POPULATION_SIZE:
        
        for i in range(int(CROSSOVER_PERCENT * len(population))):
            
            x, y = random.choice(population), random.choice(population)
            
            break_index = random.choice(list(range(10,15)))
            child1, child2 = crossover2(x, y, break_index)
            
        for i in range(int(MUTATIONS_PERCENT * len(population))):
            
            x = random.choice(population)
            
            chars = list(CHARS)
            char1 = random.choice(chars)
            chars.remove(char1)
            char2 = random.choice(chars)
            
            child3 = mutate(x, char1, char2)
            
        new_population.append(child1)
        new_population.append(child2)
        new_population.append(child3)
    return new_population
```

<div dir="rtl">

### اجرا کردن الگوریتم



</div>

```python
def run():
    population = [init_mapping() for i in range(POPULATION_SIZE)]

    last_score = 0
    last_score_increase = 0
    iterations = -STABILITY_INTERVALS

    # Run the genetic algorithm
    while last_score_increase < STABILITY_INTERVALS:
        population = generate(population)

        best_score, population = select(population)

        if (best_score > last_score):
            last_score_increase = 0
            last_score = best_score
            print("new best score is: ", last_score)
        else:
            last_score_increase += 1

        iterations += 1
    print('Best solution found after {} iterations:'.format(iterations))
    print(population[0])
```

```python
run()
```
